Uma possibilidade e excluir a coluna conforme abaixo

```python
base_credit.drop(base_credit[base_credit.age < 0].index, inplace= True)
base_credit.loc[base_credit.age < 0]
```

Uma das possibilidade e preencher os valores com a media.

```python
base_credit = pd.read_csv("credit-data.csv") #importa novamente a base
medias = base_credit.loc[base_credit.age > 0].mean() #salva media dos dados cuja idade > 0 na variavel
base_credit.loc[base_credit.age < 0,["age"]] = medias.age #subistitui idades < 0 pela medida das idades.
```

```python
base_credit.describe()
```

Outra possibilidade é tranformar o valor negativo em positivo

```python
base_credit = pd.read_csv("credit-data.csv") #importa novamente a base

base_credit["age"] = base_credit["age"].abs()
base_credit.describe()
```
